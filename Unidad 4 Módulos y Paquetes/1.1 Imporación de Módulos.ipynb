# 1.1 Importaicón de Módulos.

En Python, los módulos son archivos autocontenidos con unidades de código reutilizables, como funciones, clases y variables. Importar módulos locales permite organizar el código de forma eficaz, facilita su mantenimiento y mejora la reutilización del código. En este artículo, explicaremos cómo importar módulos locales en Python.

- Módulos de Python: Piense en ellos como archivos individuales de Python (p. ej., mimódulo.py) que contienen código reutilizable.

- Paquetes de Python: Son directorios (con un archivo \_\_init\_\_.py) que agrupan módulos relacionados, creando una estructura jerárquica.

Técnicas básicas de importación: Uso de la instrucción `import`

La sintaxis de `import` importa un módulo completo.

`import nombre_del_módulo`

O bien, para importar un atributo o función específica de un módulo:

`from nombre_del_módulo import nombre_del_atributo`
