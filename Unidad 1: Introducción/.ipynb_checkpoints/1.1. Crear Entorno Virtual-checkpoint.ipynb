# Actividad 1.1: Crear un Entorno Virtual
Antes de comenzar, es necesario crear un entorno virtual.
Un entorno virtual es un directorio con una instalacin de python independiente, esto nos ayuda a poder instalar slo los paquetes que necesitemos para nuestros proyectos.

Para crear un entorno virtual abre Anaconda Navigator. Vers una pantalla como la de la imagen.



Del lado izquierdo encontrars una pestaa llamada **Environments**, da click en ella. Se abrir la siguiente ventana.



Debajo, da click en **Create** para crear un entorno virtual nuevo.
Ingresa un nombre y elige Python junto con la versin que quieras usar.
Elige **Create** y se habr crear un entorno virtual con el nombre y versin de python que hayas elegido.
Una vez terminado asegrate que tu nuevo entorno est activo, esto lo puedes comprobar observando si el nombre de tu nuevo entorno tiene una franja verde del lado izquierdo y un botn de **play** del lado derecho.

Si no estuviera activo, basta con dar click en el nombre de tu entorno para que Anaconda lo cargue junto con todos los paquetes necesarios.




Vuelve a la pestaa de **Home** busca Jupyter Notebook. Da click en **Install**. Una vez que haya terminado de instalarse, inicia la aplicacin de Notebook dando click en **Launch**.



Si no hay problemas, una nueva pestaa debera abrirse en tu navegador. Si esto no ocurre, escribe en una nueva pestaa la siguiente direccin:
