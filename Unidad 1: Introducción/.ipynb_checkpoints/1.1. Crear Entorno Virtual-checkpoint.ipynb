# Actividad 1.1: Crear un Entorno Virtual
Antes de comenzar, es necesario crear un entorno virtual.
Un entorno virtual es un directorio con una instalación de python independiente, esto nos ayuda a poder instalar sólo los paquetes que necesitemos para nuestros proyectos.

Para crear un entorno virtual abre Anaconda Navigator. Verás una pantalla como la de la imagen.



Del lado izquierdo encontrarás una pestaña llamada **Environments**, da click en ella. Se abrirá la siguiente ventana.



Debajo, da click en **Create** para crear un entorno virtual nuevo.
Ingresa un nombre y elige Python junto con la versión que quieras usar.
Elige **Create** y se habrá creará un entorno virtual con el nombre y versión de python que hayas elegido.
Una vez terminado asegúrate que tu nuevo entorno esté activo, esto lo puedes comprobar observando si el nombre de tu nuevo entorno tiene una franja verde del lado izquierdo y un botón de **play** del lado derecho.

Si no estuviera activo, basta con dar click en el nombre de tu entorno para que Anaconda lo cargue junto con todos los paquetes necesarios.




Vuelve a la pestaña de **Home** busca Jupyter Notebook. Da click en **Install**. Una vez que haya terminado de instalarse, inicia la aplicación de Notebook dando click en **Launch**.



Si no hay problemas, una nueva pestaña debería abrirse en tu navegador. Si esto no ocurre, escribe en una nueva pestaña la siguiente dirección:

http://localhost:8888/tree?

¡Muy bien! Estás listx para comenzar a aprender a programar en *Python*.
